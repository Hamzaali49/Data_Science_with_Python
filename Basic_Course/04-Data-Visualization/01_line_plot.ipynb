## Import libraries
Seaborn automaticly install these libraries
- numpy
- scipy
- pandas
- matplotlib

For extensive study and practice: [Visit the manual](https://seaborn.pydata.org/index.html)

```python
#import libraries  #jupyter notebook have already installed these libraries and we just need to import them.
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
flr=sns.load_dataset("iris") #seaborn library have already dataset of iris that's why we don't need to download this.
flr

#draw a line plot 
sns.lineplot(x="sepal_length", y="sepal_width", data=flr) #here x is x-axis, y is y-axis and data is the data of "iris".
plt.show()
```

## - adding titles to line plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
flr=sns.load_dataset("iris") 
flr

#draw a line plot 
sns.lineplot(x="sepal_length", y="sepal_width", data=flr, linewidth=3, color='Red', ci=None)

#giving title to line plot
plt.title("Line plot for sepal length and width") 
plt.show()
```

## - adding limits to axes

```python
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
flr=sns.load_dataset("iris") 
flr

#draw a line plot 
sns.lineplot(x="sepal_length", y="sepal_width", data=flr) 

#adding limits to axes
plt.xlim(2)
plt.ylim(1)
plt.show()
```

## - set syles and labelling of axes
Types of styles
- darkgrid
- whitegrid
- dark
- white
- ticks

```python
import seaborn as sns
import matplotlib.pyplot as plt 

#load dataset
flr=sns.load_dataset("iris") 
flr

#draw a line plot 
sns.lineplot(x="sepal_length", y="sepal_width", data=flr)


#settting style
sns.set_style("dark")
sns.set_style(style=None, rc=None)
#labelling axes
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

plt.show()
```

## - size of figure

```python
import seaborn as sns
import matplotlib.pyplot as plt 

#load dataset
flr=sns.load_dataset("iris") 
flr

#figure size
plt.figure(figsize=(10,5)) #10 is the width and 5 is the height.

#draw a line plot 
sns.lineplot(x="sepal_length", y="sepal_width", data=flr) 

plt.show()   


```

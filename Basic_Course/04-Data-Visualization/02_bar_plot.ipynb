## - setting order and colors

```python
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
bt=sns.load_dataset("titanic")
bt

#draw a bar plot
sns.barplot(x="sex", y="alone", hue="who", data=bt)
            

#setting order
sns.barplot(x="sex", y="alone", hue="who", data=bt, order=["female", "male"])

#setting colors
sns.barplot(x="sex", y="alone", hue="who", data=bt, order=["female", "male"], color="grey")
#using palette
sns.barplot(x="sex", y="alone", hue="who", data=bt, order=["female", "male"], color="grey", palette="tab20b")
#saturation of palette
sns.barplot(x="sex", y="alone", hue="who", data=bt, order=["female", "male"], color="grey", palette="tab20b", saturation=2)

plt.show()
```

## - horizontal plot

```python
# to make horizontal plot we need to put numeric variable at x-axis and catagorical variable at y-axis. 
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
bt=sns.load_dataset("titanic")
bt

#draw a bar plot
sns.barplot(x="alone", y="sex", hue="who", data=bt,  
linewidth=3, errcolor="0", facecolor=(1, 1, 1, 0), edgecolor="0") 
#errcolor change the clr of error bar from black to grey to white, edge color chnage clr of bar edges from black to grey to white.
#facecolor(1,1,1,0.5) 1 shows R, 0 shows G ,0 shows B, and 0.5 shows opacity. Overall it is RGB coloring and opacity of color.
plt.show()
```

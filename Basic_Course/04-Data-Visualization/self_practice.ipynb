```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy
#self=pd.read_csv("p_data.CSV", sep=",")
self
plt.figure(figsize=(10, 5))
plt.title("Bar plot for weight")
sns.barplot(x="Gender", y="Your Weight in kg? (float)", data=self, hue="Gender", ci=None, palette="tab20b", saturation=2)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title("Bar plot for weight of both genders in different locations")
sns.barplot(x="Location", y="Your Weight in kg? (float)", data=self, order=["West", "Pakistan", "India", "East"], hue="Gender", estimator=numpy.median, palette="Accent", saturation=2)
plt.show()
```

```python

plt.figure(figsize=(10, 5))

sns.lineplot(x="Location", y="field_of_study", data=self, color="Red", linewidth=3)

plt.xlabel('Location', fontsize=18)

#sns.set_style(style=None, rc=None)

sns.set_style("darkgrid")

plt.ylabel('Field of Study', fontsize=16)

plt.title("Line plot for field of study", fontsize=20)
plt.show()
```
